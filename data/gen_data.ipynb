# generate train and test dataset

```python
import os
import numpy as np
import pandas as pd
import random
import json
import seaborn as sns
from IPython.display import clear_output
import matplotlib.pyplot as plt

# Data processing
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

from scipy import stats
import warnings
warnings.filterwarnings("ignore")

# from sklearn.metrics import plot_confusion_matrix
```

## PCA
Elements数据存在相关度，而且为了减少qubits数量，降低模型过拟合风险和训练时间。需要将高维数据进行降维处理，提取数据的主要特征分量。

```python
testset_ratio = 0.3
n_comps = 4

df = pd.read_csv("./toydata.csv")

Elem_col = [f'element {i}' for i in range(1,13)]
Prop_col = [f'property']

Elem_data = df[Elem_col].values
Prop_data = df[Prop_col].values

# scale Prop_data
scaler = StandardScaler()
Prop_data = scaler.fit_transform(Prop_data)
Elem_train, Elem_test, Prop_train, Prop_test = train_test_split(Elem_data, Prop_data, test_size=testset_ratio, random_state=42)

pca = PCA(n_components=n_comps)
Elem_train = pca.fit_transform(Elem_train)
Elem_test = pca.transform(Elem_test)
if not os.path.exists("./pca"):
    os.makedirs("./pca")

# np.save("./pca/Elem_train.npy", Elem_train)
# np.save("./pca/Prop_train.npy", Prop_train)
# np.save("./pca/Elem_test.npy", Elem_test)
# np.save("./pca/Prop_test.npy", Prop_test)
```

```python
Elem_train.shape, Prop_train.shape, Prop_train.shape, Prop_test.shape
```

```python
elem_dt = np.concatenate((Elem_train, Elem_test))
elem_df = pd.DataFrame({'element1':elem_dt[:,0],'element2':elem_dt[:,1], 'element3':elem_dt[:,2],'element4':elem_dt[:,3]})
column = elem_df.columns.tolist()
```

```python
elem_df
```

```python
sns.set(font_scale=2.4)

fig = plt.figure(figsize = (40, 5), dpi=100)
for i in range(n_comps):
    plt.subplot(1, 4, i+1)
    sns.boxplot(elem_df[column[i]], width=0.3)
    plt.ylabel(column[i],fontsize = 40)
    plt.xlabel(column[i],fontsize = 40)

plt.show()
```

```python
cols = 4
rows = 1
plt.figure(figsize=(4*cols, 4*rows))
# fig = plt.figure(figsize=(32, 8), dpi=100)
sns.set(font_scale=1.6)
for col in range(4):
    ax = plt.subplot(1, 4, col+1)
    sns.distplot(elem_df[column[col]], fit = stats.norm)
plt.tight_layout()
plt.show()
```
